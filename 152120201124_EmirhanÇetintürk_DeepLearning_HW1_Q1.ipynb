**Required libraries**

```python
import numpy as np

import matplotlib.pyplot as plt
from skimage import color
from skimage import io
from PIL import Image
```

**Loading images and show both of them**

```python
Image1 = plt.imread('/content/tiger.jpg')
Image2 = plt.imread('/content/tiger2.jpg')
```

```python
imgplot = plt.imshow(Image1)
plt.show()
```

```python
imgplot = plt.imshow(Image2)
plt.show()
```

**Resized both images**

```python
from skimage.transform import rescale, resize, downscale_local_mean
Image1Resized = resize(Image1, (256, 256))

imgplot = plt.imshow(Image1Resized)
plt.show()
```

```python
from skimage.transform import rescale, resize, downscale_local_mean
Image2Resized = resize(Image2, (256, 256))

imgplot = plt.imshow(Image2Resized)
plt.show()
```

**Vertical and horizontal reflections were made on images**

```python
import cv2
```

```python
Image1Vertical = Image1Resized
Image1Vertical = cv2.flip(Image1Resized, 0)
plt.imshow(Image1Vertical)
plt.show()
```

```python
Image2Vertical = Image2Resized
Image2Vertical = cv2.flip(Image2Resized, 0)
plt.imshow(Image2Vertical)
plt.show()
```

```python
Image1Horizontal = Image1Resized
Image1Horizontal = cv2.flip(Image1Resized, 1)
plt.imshow(Image1Horizontal)
plt.show()
```

```python
Image2Horizontal = Image2Resized
Image2Horizontal = cv2.flip(Image2Resized, 1)
plt.imshow(Image2Horizontal)
plt.show()
```

**Rotating both images**

```python
ResizedImage1Rotated90DegreeLeft = Image1Resized
ResizedImage1Rotated90DegreeLeft = cv2.rotate(Image1Resized, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(ResizedImage1Rotated90DegreeLeft)
plt.show()
```

```python
ResizedImage2Rotated90DegreeLeft = Image2Resized
ResizedImage2Rotated90DegreeLeft = cv2.rotate(Image2Resized, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(ResizedImage2Rotated90DegreeLeft)
plt.show
```

```python
ResizedImage1Rotated90DegreeRight = Image1Resized
ResizedImage1Rotated90DegreeRight = cv2.rotate(Image1Resized, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(ResizedImage1Rotated90DegreeRight)
plt.show
```

```python
ResizedImage2Rotated90DegreeRight = Image2Resized
ResizedImage2Rotated90DegreeRight = cv2.rotate(Image2Resized, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(ResizedImage2Rotated90DegreeRight)
plt.show
```

**Resized both images to half by keeping aspect ratio**

```python
Image1HalfResized = resize(Image1Resized, (int(256/2), int(256/2)))
plt.imshow(Image1HalfResized)
plt.show
```

```python
Image2HalfResized = resize(Image2Resized, (int(256/2), int(256/2)))
plt.imshow(Image2HalfResized)
plt.show
```

**The left part of the first image and the right part of the second image were cropped.**

```python
Image1ResizedLeftHalfCropped = Image1HalfResized[0: 128, 0: 64]
plt.imshow(Image1ResizedLeftHalfCropped)
plt.show
```

```python
Image2ResizedRightHalfCropped = Image2HalfResized[0: 128, 64: 128]
plt.imshow(Image2ResizedRightHalfCropped)
plt.show
```

**Cropped images were merged.**

```python
MergeBothImages = cv2.hconcat([Image1ResizedLeftHalfCropped, Image2ResizedRightHalfCropped])
plt.imshow(MergeBothImages)
plt.show
```

**The initial version and the combined version of the images were showed.**

```python
plt.imshow(Image1Resized)
plt.figure()
plt.imshow(Image2Resized)
plt.figure()
plt.imshow(MergeBothImages)
plt.show
```
