**Required libraries**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
from PIL import Image
```

**Loading, showing and resizing image**

```python
Image1 = plt.imread('/content/redx_best.png')
print(Image1.shape)
imgplot = plt.imshow(Image1)
plt.show()

Image1 = cv2.resize(Image1, (256,256))

cv2.imwrite('/content/Image1NoBlueChannel.png',Image1)
imgplot = plt.imshow(Image1)
plt.show()
```

**Changing color mode as RGB and setting blue and green channels to zero**


```python
Image1 = cv2.cvtColor(Image1, cv2.COLOR_BGR2RGB)
imgplot = plt.imshow(Image1)
plt.show()
```

```python
Image1[:,:,1] = np.zeros([Image1.shape[0], Image1.shape[1]])
Image1[:,:,2] = np.zeros([Image1.shape[0], Image1.shape[1]])

cv2.imwrite('/content/Image1NoBlueChannel2.png',Image1)
imgplot = plt.imshow(Image1)
plt.show()
```

**Converting image to grayscale**

```python
Image1Temp = plt.imread("/content/Image1NoBlueChannel2.png")
Image1 = cv2.cvtColor(Image1Temp, cv2.COLOR_RGB2GRAY)
cv2.imwrite('/content/Image1NoBlueChannel3.png',Image1)
plt.imshow(Image1)
plt.show()
#print(Image1.shape)
```

**Appyling gaussing filter to the image as a first derivative**

```python
from scipy.ndimage import gaussian_filter, laplace

Image1 = gaussian_filter(Image1, sigma=1)

plt.imshow(Image1)
plt.show()
```

**Appyling laplace filter to the image as a second derivative**

```python
Image1 = laplace(Image1, mode='constant')

plt.imshow(Image1)
plt.show()
```
